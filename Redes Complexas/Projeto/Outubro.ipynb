```python
import pandas as pd

# 1. Defina o nome do seu arquivo de entrada e do novo arquivo
nome_arquivo_entrada = '../P1/dataset_zap_1.csv'
nome_arquivo_saida = 'dados_outubro.csv'
nome_coluna_data = 'date_message'

try:
    # 2. Ler o arquivo CSV
    # O parâmetro parse_dates é crucial para que o pandas reconheça a coluna como data
    df = pd.read_csv(nome_arquivo_entrada, parse_dates=[nome_coluna_data])
    
    print(f"Arquivo '{nome_arquivo_entrada}' lido com sucesso. Total de linhas: {len(df)}")
    # 3. Filtrar o DataFrame para apenas as linhas onde o mês da coluna de data é Outubro (10)
    # Acessamos o atributo .dt.month da coluna de data para obter o número do mês
    df_outubro = df[df[nome_coluna_data].dt.month == 10]
    
    print(f"Filtragem concluída. Total de linhas em Outubro: {len(df_outubro)}")
    
    # 4. Salvar o DataFrame filtrado em um novo arquivo CSV
    # index=False evita que o pandas salve o índice do DataFrame como uma coluna no novo CSV
    df_outubro.to_csv(nome_arquivo_saida, index=False)
    
    print(f"Dados de Outubro salvos com sucesso no arquivo '{nome_arquivo_saida}'.")

except FileNotFoundError:
    print(f"ERRO: O arquivo '{nome_arquivo_entrada}' não foi encontrado. Verifique o nome e o caminho.")
except KeyError:
    print(f"ERRO: A coluna de data '{nome_coluna_data}' não foi encontrada no arquivo CSV. Verifique se o nome está correto.")
except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")
```
